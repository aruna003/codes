7. Use the DataFrame created in Question 5 above to do the following:  
a) Append the DataFrame Sales2 to the DataFrame Sales.  
b) Change the DataFrame Sales such that it becomes its transpose.  
c) Display the sales made by all sales persons in the year 2017.  
d) Display the sales made by Madhu and Ankit in the year 2017 and 2018.  
e) Display the sales made by Shruti 2016.  
f) Add data to Sales for salesman Sumeet where the sales made are [196.2, 37800, 52000, 78438, 38852] in the years [2014, 2015, 2016, 2017, 2018 respectively.  
g) Delete the data for the year 2014 from the DataFrame Sales.  
h) Delete the data for sales man Kinshuk from the DataFrame Sales.  
i) Change the name of the salesperson Ankit to Vivaan and Madhu to Shailesh.   
j) Update the sale made by Shailesh in 2018 to 100000.  
k) Write the values of DataFrame Sales to a comma separated file SalesFigures.csv on the disk. Do not write the row labels and column labels.   
l) Read the data in the file SalesFigures.csv into a DataFrame SalesRetrieved and Display it. Now update the row labels and column labels of SalesRetrieved to be the same as that of Sales

```python
data = {
    '2014': [100.5, 150.8, 200.9, 30000],
    '2015': [12000, 18000, 22000, 30000],
    '2016': [20000, 50000, 70000, 100000]
}

index = ['Madhu', 'Kusum', 'Kinshuk', 'Ankit']

Sales = pd.DataFrame(data, index=index)

data2 = {
    '2018':[160000 ,110000 ,500000 ,34000]
}
index_val = ["Madhu","Kusum","Kinshuk","Ankit"]
Sales2 = pd.DataFrame(data2,index=index_val)

pd.concat([Sales,Sales2])
print("Appended:\n",Sales)

Sales = Sales.transpose()
print("Transpose:\n",Sales)

print("2016:",Sales.loc["2016"])

sales_madhu_ankit = Sales.loc[['2016'],['Madhu', 'Ankit']]
print("Madhu and Ankit:\n",sales_madhu_ankit)

sales_shruti_2016 = Sales.loc[['2016'],['Madhu']]
print("Shruti 2016:\n",sales_shruti_2016)

Sales['Sumeet'] = [196.2, 37800, 52000]
print("Col added:\n",Sales)

Sales = Sales.drop('2014', axis=0)
print("Drop 2014:\n",Sales)

Sales = Sales.drop('Kinshuk', axis=1)
print("Drop Kinush:\n",Sales)

Sales = Sales.rename(columns = {'Ankit': 'Vivaan', 'Madhu': 'Shailesh'})
print("Renamed:\n",Sales)

Sales.at['2018', 'Shailesh'] = 100000
print("Sales updated:\n",Sales)

Sales.to_csv('SalesFigures.csv', header=False, index=False)

SalesRetrieved = pd.read_csv('SalesFigures.csv', header=None)
SalesRetrieved.index = Sales.index
SalesRetrieved.columns = Sales.columns
print("SalesRetrieved:\n",SalesRetrieved)

```
